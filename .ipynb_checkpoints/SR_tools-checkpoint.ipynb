```python
from pathlib import Path
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
datapath = Path("/home/carloscano/data/SR")
```

```python
folder_list = [folder for folder in datapath.glob("*")]
```

```python
for folder in folder_list:
    file_list = [file for file in folder.glob("*")]
    df_all = pd.DataFrame()
    for file in file_list:
        names_split =  file.name.split("_")
        day = datetime(year=int(names_split[2]), month=int(names_split[3]), day=int(names_split[4]))
        df = pd.read_table(file).T
        df.columns = df.iloc[0]
        df = df.drop("-10")
        df.index = [day + pd.Timedelta(hours = int(index) -1) for index in df.index.to_numpy()]
        Path.joinpath(folder, "/process/"+file.name)
        df.to_csv(Path.joinpath(folder, "/process/"+day.strftime("%y_%m_%d"), compression="xz")
        break
    break
```
